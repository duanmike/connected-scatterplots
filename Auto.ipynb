```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas import read_csv

df = read_csv('AutoData.csv')
data = df.values
df.head(5)
```

```python
df.tail(5)
```

```python
df.plot(kind='line',x='Year',y='Miles_Capita')
df.plot(kind='line',x='Year',y='Deaths_100K')
plt.show
```

```python
fig,ax = plt.subplots()
ax.plot(df.Year, df.Deaths_100K, color="green", marker="o", markerfacecolor='white', markersize=3)
ax.set_xlabel("Year")
ax.set_ylabel("Auto Fatalities per 100K People", color="green")

ax2=ax.twinx()
ax2.plot(df.Year, df.Miles_Capita, color="blue", marker="o", markerfacecolor='white', markersize=3)
ax2.set_ylabel("Miles Driven Per Capita", color="blue")

plt.title("Auto Fatalities and Miles Driven per Year")
plt.savefig('auto_dalc', dpi=300, bbox_inches="tight")
plt.show
```

```python
fig,ax = plt.subplots()
ax.plot(df.Miles_Capita, df.Deaths_100K, color="purple", marker="o", markerfacecolor='white', markersize=3)
ax.set_xlabel("Miles Driven Per Capita")
ax.set_ylabel("Auto Fatalities per 100K People")
ax.set_title("Miles Driven vs. Auto Fatalities (1970 - 2020)")

plt.annotate( 'Direction (Time)', xy=(7500,21),
              xytext=(8000, 21.5),
              arrowprops=dict( arrowstyle="<-" )
            )

plt.savefig('auto_cs', dpi=300)
plt.show
```
